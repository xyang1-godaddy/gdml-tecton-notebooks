```pyspark
from pyspark.sql.functions import explode, size

df = spark.sql("select * from tecton_dev.uds_traffic_session")

df = df.filter(df.shopper_ids.isNotNull() & (size(df.shopper_ids) > 0)).withColumn(
    "shopper_id", explode("shopper_ids")
)
```

```pyspark
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window

window = Window.partitionBy("shopper_id").orderBy(col("session_begin_date").desc())
df = df.withColumn("row_number", row_number().over(window)).filter(col("row_number") <= 10)

df = df.drop("row_number")
```

```pyspark
from pyspark.sql.functions import col, row_number, explode, struct, collect_list
from pyspark.sql.window import Window

page_name_count_df = df.withColumn("page_name", explode("fully_qualified_page_names")).groupBy("shopper_id", "page_name").count()
host_name_count_df = df.withColumn("host_name", explode("host_names")).groupBy("shopper_id", "host_name").count()

window = Window.partitionBy("shopper_id").orderBy(col("count").desc())
page_name_count_df = page_name_count_df.withColumn("row_number", row_number().over(window)).filter(col("row_number") <= 10)
host_name_count_df = host_name_count_df.withColumn("row_number", row_number().over(window)).filter(col("row_number") <= 10)

page_name_count_df = page_name_count_df.select("shopper_id", struct('page_name', 'count').alias("name_count_struct"))
host_name_count_df = host_name_count_df.select("shopper_id", struct('host_name', 'count').alias("name_count_struct"))

page_name_count_df = page_name_count_df.groupBy("shopper_id").agg(collect_list("name_count_struct").alias("top_pages"))
host_name_count_df = host_name_count_df.groupBy("shopper_id").agg(collect_list("name_count_struct").alias("top_hosts"))


page_name_count_df = page_name_count_df.withColumn("top_pages_json", to_json("top_pages"))
host_name_count_df = host_name_count_df.withColumn("top_hosts_json", to_json("top_hosts"))

result = page_name_count_df.join(host_name_count_df, "shopper_id")
```
